```python
from sklearn.model_selection import train_test_split

feature_names = ['Tenure', 'Usage Frequency', 
        'Age_low', 'Age_middle', 'Age_high', 'Gender_male',
        'Many Support Calls', 'Many Payment Delay', 'Long_contract',
        'Many spend', 'Interaction Group']

X = data_train[feature_names]
y = data_train['Churn']

X_train, X_valid, y_train, y_valid = train_test_split(
    X, y, test_size = 0.25, random_state=1234)
```

## ëª¨ë¸ êµ¬í˜„ ë° ì„±ëŠ¥ í‰ê°€

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

estimators = [DecisionTreeClassifier(class_weight='balanced', random_state=1234),
             RandomForestClassifier(class_weight='balanced', random_state=1234),
             GradientBoostingClassifier(random_state=1234)]
estimators
```

```python
# ëª¨ë¸ êµ¬í˜„ ë° ì„±ëŠ¥ í‰ê°€
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report, f1_score, accuracy_score, roc_auc_score

# ë„ˆë¬´ ì˜¤ë˜ê±¸ë¦¬ë¯€ë¡œ ìƒ˜í”Œë§
X_small = X_train.sample(frac=0.3, random_state=1234)
y_small = y_train.loc[X_small.index]

results = []

for estimator in estimators:
    result = []
    if estimator.__class__.__name__ == 'DecisionTreeClassifier':
        param_distributions = {
            'max_depth': [None, 5, 10, 15],
            'max_features': [0.5, 0.7, 1.0]
        }
    else:
        param_distributions = {
            'max_depth': [None, 5, 10, 15],
            'max_features': [0.5, 0.7, 1.0],
            'n_estimators': [100, 150, 200]
        }

    clf = RandomizedSearchCV(estimator,
                       param_distributions=param_distributions,
                       n_iter=10,
                       scoring='roc_auc',  # AUC
                       n_jobs=-1,
                       cv=3)
    


    clf.fit(X_small, y_small)

    best_model = clf.best_estimator_
    
    pred_valid = best_model.predict(X_valid)
    valid_f1 = f1_score(y_valid, pred_valid)
    
    result.append(estimator.__class__.__name__)
    result.append(clf.best_params_)
    result.append(clf.best_score_)  # CV ì ìˆ˜
    result.append(valid_f1)         # validation ì ìˆ˜
    result.append(best_model) 
    results.append(result)
```

```python
# best model ì°¾ê¸°
df = pd.DataFrame(results, columns=['estimator', 'best_params','cv_score','valid_f1','model'])

dfdf = pd.DataFrame(df).sort_values(by='valid_f1', ascending=False).head()
dfdf
```
